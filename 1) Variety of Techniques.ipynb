# Grouping the data â€“ aggregation, filtering, and transformation

```python
### create a simple dummy data frame that has a rich mix of both numerical and categorical variables
a=['Male','Female']
b=['Rich','Poor','Middle Class']
gender=[]
seb=[]
for i in range(1,101):
    gender.append(np.random.choice(a))
    seb.append(np.random.choice(b))
height=30*np.random.randn(100)+155
weight=20*np.random.randn(100)+60
age=10*np.random.randn(100)+35
income=1500*np.random.randn(100)+15000

df=pd.DataFrame({'Gender':gender,'Height':height,'Weight':weight,'Age':age,'Income':income,'Socio-Eco':seb})
df.head()
```

```python
### to group the data frame based on the Gender variable
grouped = df.groupby('Gender')
grouped.groups
```

```python
grouped=df.groupby('Gender')
for names,groups in grouped:
    print (names)
    print (groups)
```

```python
### A single group can be selected by writing the following:
grouped.get_group('Female')
```

```python
grouped=df.groupby(['Gender','Socio-Eco'])
for names,groups in grouped:
    print (names)
    print (groups)
```

```python
# sum function
grouped=df.groupby(['Gender','Socio-Eco'])
grouped.sum()
```

```python
### To get the number of rows in each group (or calculate the size of each group)
grouped=df.groupby(['Gender','Socio-Eco'])
grouped.size()
```

```python
### to get the summary statistics for each group separately
grouped=df.groupby(['Gender','Socio-Eco'])
grouped.describe()
```

```python
### calculate sum of Income, mean of Age, and standard deviation of Height
grouped=df.groupby(['Gender','Socio-Eco'])
grouped.aggregate({'Income':np.sum,'Age':np.mean,'Height':np.std})
```

```python
grouped=df.groupby(['Gender','Socio-Eco'])
grouped.aggregate({'Age':np.mean,'Height':lambda
                   x:np.mean(x)/np.std(x)})
```

```python
### apply several functions to all the columns at the same time
grouped.aggregate([np.sum, np.mean, np.std])
```

```python
### choose elements from the Age column that are a part of the group wherein the sum of Age is greater than 700
groupage = grouped['Age'].filter(lambda x:x.sum()>700)
groupage.head()
```
